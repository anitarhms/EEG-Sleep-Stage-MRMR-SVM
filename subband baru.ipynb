```python
#membuat fungsi butterworth bandpass filter order 5
def butter_bandpass(lowcut, highcut, fs, order=5): 
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a
#fungsi u/ mengaplikasikan filter ke data 
def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data) #lfilter itu forward filtering
    return y
#inisiasi nilai fs, range
fs = 100.0
lowcut = 0.5
highcut = 45.0 

fpz_new = butter_bandpass_filter(x, lowcut, highcut, fs, order=5) #bandpass 
```

```python
print(fpz_new)
```

```python
#fungsi dekomposisi wavelet
for i in range (len(fpz_new)) :
  wp = pywt.WaveletPacket(data=fpz_new[i], wavelet='db2', mode='symmetric',maxlevel=7) #dekomposisi waveletpaket daubechies 2
  
  #6 subpita
  sinyal_delta=np.concatenate([wp['aaaaaad'].data, wp['aaaaad'].data,wp['aaaada'].data,wp['aaaadd'].data,wp['aaadaaa'].data])
  sinyal_theta=np.concatenate([wp['aaadaad'].data, wp['aaadad'].data,wp['aaadd'].data,wp['aadaa'].data,wp['aadad'].data])
  sinyal_alfa=np.concatenate([wp['aadda'].data, wp['aaddda'].data])
  sinyal_spindle=np.concatenate([wp['aaddd'].data, wp['adaaa'].data,wp['adaad'].data])
  sinyal_beta1=np.concatenate([wp['adad'].data, wp['adda'].data])
  sinyal_beta2=np.concatenate([wp['addd'].data, wp['da'].data])
  
  #mean quadratic value
  f1=np.sqrt(np.mean(sinyal_delta**2))
  f2=np.sqrt(np.mean(sinyal_theta**2))
  f3=np.sqrt(np.mean(sinyal_alfa**2))
  f4=np.sqrt(np.mean(sinyal_spindle**2))
  f5=np.sqrt(np.mean(sinyal_beta1**2))
  f6=np.sqrt(np.mean(sinyal_beta2**2))
  
  #total 
  f7=np.sum([f1,f2,f3,f4,f5,f6])
  #Mean of the absolute values of the coefficients in each subband 
  f8=np.mean(np.abs(sinyal_delta))
  f9=np.mean(np.abs(sinyal_theta))
  f10=np.mean(np.abs(sinyal_alfa))
  f11=np.mean(np.abs(sinyal_spindle))
  f12=np.mean(np.abs(sinyal_beta1))
  f13=np.mean(np.abs(sinyal_beta2))
  #Standard deviation of the coefficients in each subband 
  f14=np.std(sinyal_delta)
  f15=np.std(sinyal_theta)
  f16=np.std(sinyal_alfa)
  f17=np.std(sinyal_spindle)
  f18=np.std(sinyal_beta1)
  f19=np.std(sinyal_beta2)
  #Ratio of different mean absolute values in different subbands 
  f20=f8/f9
  f21=f9/f10
  f22=f10/f11
  f23=f11/f12
  f24=f12/f13
  #Shannon entropy 
  vectorB=([sinyal_delta,sinyal_theta,sinyal_alfa,sinyal_spindle,sinyal_beta1,sinyal_beta2])
  vb_series = pd.Series(vectorB)
  counts = vb_series.value_counts()
  f25 = entropy(counts)
  #Permutation entropy 
  prm=np.concatenate([sinyal_delta,sinyal_theta,sinyal_alfa,sinyal_spindle,sinyal_beta1,sinyal_beta2])
  f26=ent.permutation_entropy(prm, normalize=True)
  print(i,f1,f2,f3,f4,f5,f6,f7,f8,f9,f10,f11,f12,f13,f14,f15,f16,f17,f18,f19,f20,f21,f22,f23,f24,f25,f26)
```
