```python
import pandas as pd
import numpy as np
#library untuk mrmr
from sklearn.feature_selection import f_regression
from sklearn.model_selection  import train_test_split
#svm + bootstrap
from sklearn.utils import resample
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV #tuning param
from sklearn.metrics import classification_report, confusion_matrix  
#plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os
import glob
```

```python
fitur = pd.DataFrame([])  #membuat dataframe 
i=1
for i in range(11): #looping 
  i=i+1
  FILE_PATH = '/content/drive/MyDrive/csv fitur/subband baru/'+str(i)+'.csv' #path file csv dalam satu folder, pakai looping for i in range agar bacanya urut 
  data = pd.read_csv(FILE_PATH, sep=",", header=0) #read csv, berdasarkan path, row 0 sbg header 
  
  fitur = pd.concat([fitur,data],axis=0) #concatenate csv cs jadi satu data frame 'all_data'
```

```python
pd_data=pd.DataFrame(fitur)
print(pd_data)
```

```python
target=pd_data.iloc[:,0]
y=pd.Series(target)
print(y)
```

```python
fitur = pd_data.drop("class", axis=1)
X=pd.DataFrame(fitur)
```

```python
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) #split data 80:20
```

```python
F=pd.Series(f_regression(X_train,y_train)[0],index=X_train.columns) #menghitung f regression 
corr=X_train.corr().abs().clip(.00001) #menghitung korelasi, .clip biar ga divide by 0 

#buat list selected dan ga 
selected=[]
not_selected=list(X_train.columns)

for i in range(15):
  score=F.loc[not_selected]/corr.loc[not_selected,selected].mean(axis=1).fillna(.00001) #FCQ = F/corr
  print(score) #print score mrmr
  best=score.index[score.argmax()] #memilih yg paling tinggi 
  selected.append(best) #dimasukian ke list selected + digabung 
  not_selected.remove(best) #menghilangkan fitur yg sudah terpilih
  
```

```python
#fitur diskriminan
 
n_iterations = 10  #No. of bootstrap samples to be repeated (created)
n_size = int(len(pd_data) * 0.80)
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
print(n_size)
```

```python
dat=pd_data.drop(['f1', 'f3', 'f5', 'f7', 'f8', 'f9', 'f10', 'f14', 'f15', 'f16', 'f18', 'f20', 'f21', 'f25', 'f27', 'f31', 'f32'],axis=1)
```

```python
values=dat.values
print(values)
```

```python
#tuning parameter
svc = SVC(kernel='rbf',decision_function_shape='ovo')
grid = GridSearchCV(svc, param_grid,refit = True, verbose = 3)
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
print(grid.score(X_test,y_test))
```

```python
#https://www.kaggle.com/sasha18/resampling-methods-using-bootstrap-cv
stats = list()
for i in range(n_iterations):

    #prepare train & test sets
    train = resample(values, n_samples = n_size) #Sampling with replacement..whichever is not used in training data will be used in test data
    test = np.array([x for x in values if x.tolist() not in train.tolist()]) #np.array([x for x in values if x not in train]) #picking rest of the data not considered in training sample
    
    #fit model
    model = SVC(kernel='rbf',decision_function_shape='ovo',C=1000,gamma=0.0001)
    model.fit(train[:,1:], train[:,0]) #model.fit(X_train,y_train) i.e model.fit(train set, train label as it is a classifier)
    
    #evaluate model
    predictions = model.predict(test[:,1:]) #model.predict(X_test)
    score = accuracy_score(test[:,0], predictions) #accuracy_score(y_test, y_pred)
    #caution, overall accuracy score can mislead when classes are imbalanced
    
    print(score)
    stats.append(score)
```

```python
accuracy=np.array(stats)
print('average: ', accuracy.mean()*100)
```

```python
cnfmatrix=confusion_matrix(test[:,0],  predictions)
print(cnfmatrix)
```

```python
n_classes=3
for c in range(n_classes):
    tp = cnfmatrix[c,c]
    fp = sum(cnfmatrix[:,c]) - cnfmatrix[c,c]
    fn = sum(cnfmatrix[c,:]) - cnfmatrix[c,c]
    tn = sum(np.delete(sum(cnfmatrix)-cnfmatrix[c,:],c))

    sensivity = tp/(tp+fn)
    specificity = tn/(tn+fp) 

    print("for class {}: sensivity {}, specificity {}, tp {}, fp {},fn {},tn {}".format(c,round(sensivity,4), round(specificity,4),round(tp),round(fp),round(fn),round(tn)))
```
